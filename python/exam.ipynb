```python
import numpy as np

a = np.array([1, 2])
b = np.array([3, 4])

dot_product = np.dot(a, b)

print('a) ', dot_product)

a = np.array([1, 3, 4])
b = np.array([0, 2, -3])

dot_product = np.dot(a, b)

print('b) ', dot_product)
```
